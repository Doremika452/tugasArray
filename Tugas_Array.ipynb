1) Mendefinisikan sebuah list kosong untuk menyimpan `value` Mahasiswa

```python
value=[]
```

2. Melakukan import `matplotlib.pyplot as plt` untuk menunjukkan grafik

```python
import matplotlib.pyplot as plt
```

3. Menggunakan `for` sebagai perulangan untuk meminta input `value` Mahasiswa dari pengguna sebanyak 10 kali dan menampilkannya menggunakan `print("Nilai Mahasiswa : ", nilai)`

```python
for i in range(10):
    angka = int(input("Masukkan Nilai Mahasiswa : "))
    value.append(angka)

print("Nilai Mahasiswa : ", value)
```


4. Menampilkan nilai tertinggi dan terendah mahasiswa menggunakan `max(value)` dan `min(value)`

```python
print("Nilai Tertinggi : ", max(value))
print("Nilai Terendah : ", min(value))
```

5. Menampilkan nilai rata-rata mahasiswa menggunakan sum(value)/len(value)

    `sum(value)` untuk `menjumlahkan nilai-nilai` pada array

    `len(value)` untuk `menampilkan total` nilai pada array

```python
print("Nilai Rata-rata : ", sum(value)/len(value))
```

6. Menggunakan `for` untuk perulangan pada elemen dalam `array/list nilai`

    mengecek kondisi dengan `if` , jika kondisi >= 60 maka lulus += 1

    Menghitung `tidak_lulus` dengan mengurangi total nilai `( len(nilai))` dengan `total lulus`

    `jumlah mahasiswa` yang lulus dan tidak lulus ditampilkan dengan menggunakan `print`

```python
lulus = 0
for n in value:
    if n >= 60:
        lulus += 1
tidak_lulus = len(value) - lulus

print("Jumlah Mahasiswa yang lulus : ", lulus)
print("Jumlah Mahasiswa yang tidak lulus : ", tidak_lulus)
```

7. Untuk menampilkan grafik:

    Grafik yang menggambarkan nilai tertinggi dan terendah

    Grafik yang menggambarkan data kelulusan Mahasiswa

```python
plt.bar(["Terendah", "Tertinggi"], [min(value), max(value), ])
plt.title("Grafik Nilai Mahasiswa")
plt.ylabel("Nilai")
plt.ylim([0, 100])
plt.show()

print("\n")

plt.bar(["Lulus", "Tidak Lulus"], [lulus, tidak_lulus] )
plt.title("Grafik Kelulusan Mahasiswa")
plt.ylabel("Mahasisawa")
plt.ylim([0, 10])
plt.show()
```
